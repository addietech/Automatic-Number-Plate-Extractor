```python
import easyocr
import csv

# Define the OCR reader for the English language
reader = easyocr.Reader(['en'])

# Open the preprocessed number plate image
number_plate_image_path = "/content/demo_image.jpg"

# Use EasyOCR to perform OCR on the image
result = reader.readtext(number_plate_image_path)

# Extract recognized text from the result
if result:
    plate_text = result[0][1]
else:
    plate_text = "Unable to recognize"

# Specify the CSV file path
csv_file_path = "number_plate_data.csv"

# Write the data to the CSV file
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Number Plate", "Plate Text"])
    writer.writerow(["1", plate_text])

print("Number Plate Text:", plate_text)
```
